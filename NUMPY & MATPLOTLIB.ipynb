```python
# define variables
t = np.arange(0, 10, 0.05)
freq = 1.0

# define function
y = np.sin(2*np.pi*freq*t)
shifted_y = np.sin(2*np.pi*freq*t - 2.0)

# plotting the figure
fig, ax = plt.subplots()
ax.plot(t, y)
ax.plot(t, shifted_y)
```

## BASIC NUMPY/MATPLOTLIB EXERCISES

```python
# NUMBER 1 

# defining functions
x = np.arange(0, 1, 0.000001)

f1 = 3.0
f2 = 1.0

sin_1 = np.sin(2*np.pi*f1*x)
sin_2 = np.sin(2*np.pi*f2*x)
prod = sin_1 * sin_2

fig, ax = plt.subplots()
ax.plot(x, prod, label="product")
ax.plot(x, sin_1, label="sine 1")
ax.plot(x, sin_2, label="sine 2")
ax.legend()
```

```python
# NUMBER 2 

# grid
x = np.linspace(0, 1, 100)

# equation of quarter circle
y = np.sqrt(1-x**2)

#plotting the figure
fig, ax = plt.subplots()
ax.plot(x, y, label="Quarter Circle")
ax.legend()
```

```python
# NUM 2 PART 2

# grid
fig, ax = plt.subplots()
x = np.linspace(0, 1, 100)


# equation of quarter circle
y = np.sqrt(1-x**2)
inside = 0
out = 0

for i in range(1000):
    x_i = np.random.random()
    y_i = np.random.random()
    if (x_i**2 + y_i**2 < 1):
        ax.scatter(x_i, y_i)
        inside += 1
    else:
        out += 1

#plotting the figure
print("inside =", inside)
print("out =", out)
ax.plot(x, y, label="Quarter Circle")
ax.legend()
```

```python
# NUMBER 2 FASTER WAY
fig, ax = plt.subplots()
N = 10000000

x_i = np.random.random(size = N)
y_i = np.random.random(size = N)
is_inside = x_i**2 + y_i**2 < 1

ax.scatter(x_i[is_inside], y_i[is_inside])
print(4*np.sum(is_inside)/N)
```

```python
# NUMBER 3

fig, ax = plt.subplots()

N = 10000
weibull = np.random.weibull(5, size=N)
ax.hist(weibull, bins=np.linspace(-4, 4, 100), density=True)
```

```python
# NUMBER 3

fig, ax = plt.subplots()

N = 100000
vonmises = np.random.vonmises(5, 100, size=N)
ax.hist(vonmises, bins=np.linspace(-4, 4, 100), density=True)
```
